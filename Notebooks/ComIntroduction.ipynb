# 通訊系統簡介

## 何謂通訊(Communication)

將訊息由某地(或某人)傳遞給某地(另一人)的過程，稱之為通訊(Communication)。  
通訊理論的主要觀點在於將信號(signal)完整地傳遞、接收並正確地解讀出來。其作法有：
 
- 加大信號強度(signal strength)以抵禦雜訊干擾 。
- 改由較乾淨的傳輸通道(transmission channel)傳送信號。
- 更改信號波形使其能適合於目前通道上傳輸。 
- 改用不易受干擾的傳輸信號，如以光取代電。

![signal transform](Figures/signal_transform.png)  
基本上我們將所要傳遞的訊息稱之為訊息(message)，而訊息可分成類比(analog)和數位(digital)訊息；以類比信號進行調變/解調變之系統謂之類比通訊系統，以數位信號進行調變/解調變之系統謂之數位通訊系統。

## 調變/解調變
這種將原始信號轉換成更適合於傳輸介質(channel)，藉以提高傳送距離及傳輸效率的方式稱之為調變(modulation)，而接收端的相反轉換方式可以使波形轉變成原信號的方式稱之為解調(demodulation)。  
![modulation/demodulation](Figures/mod_dem.png)  

一般傳輸介質皆有適合傳輸的頻帶，不一定適合原始訊息直接傳送，須藉由載波信號將訊息移往適合的頻段進行傳送，即所謂的載波調變。  
載波信號一般多為高頻信號(10kHz～1000GHz)，有時稱之為射頻(Radio Frequency；RF)信號，經常用正弦波代表高頻載波，如下列例子：  
$c(t)=V_p \cos(2\pi f t +\theta) $

1. 若$V_p$是訊息(調變信號)的函數：稱此種調變方式為振幅調變(Amplitude Modulation；AM)。 
2. 若$f$是訊息的函數：即稱此種調變方式為頻率調變(Frequency Modulation；FM)。 
3. 若相角$\theta$隨訊息而變：則稱此種調變方式為相位調變(Phase Modulation；PM)。 

```octave
t=0:0.001:2;
m=0.2*cos(2*pi*5*t);
c=cos(2*pi*100*t);
plot(t(1:400),m(1:400),t(1:400),c(1:400))
axis([0 0.4 -1.5 1.5]); xlabel('t'); title('carrier & message signals')
legend('message', 'carrier')
```

```octave
am=(1+m).*c; pm=cos(2*pi*100*t+30*m); fm=cos(2*pi*100*t+10*sin(2*pi*5*t));
subplot(3,1,1); plot(t(1:400),m(1:400),t(1:400),am(1:400)); axis([0 0.4 -1.5 1.5]); title('AM signal')
subplot(3,1,2); plot(t(1:400),m(1:400),t(1:400),pm(1:400)); axis([0 0.4 -1.5 1.5]); title('PM signal')
subplot(3,1,3); plot(t(1:400),m(1:400),t(1:400),fm(1:400)); axis([0 0.4 -1.5 1.5]); title('FM signal'); xlabel('t')
```

### 通道效應

在信號傳輸的過程中，時常會出現一些不必要的效應以致於影響信號的正確接收：
- 衰減(decay)。
- 失真(distortion)。
- 雜訊(noise)。
- 干擾(disturbance)。

一般而言，雜訊是累加在信號之上，失真則會造成信號波形的“部分變形”。所以雜訊是相加性、線性的。
